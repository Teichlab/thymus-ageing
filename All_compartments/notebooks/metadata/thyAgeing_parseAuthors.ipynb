```python
import os, sys
import pandas as pd 
import numpy as np

# Add repo path to sys path (allows to access scripts and metadata from repo)
repo_path,_ = os.path.split(os.path.split(os.getcwd())[0])
repo_path = '/nfs/team205/lm25/thymus_projects/thymus_ageing_atlas/General_analysis'
sys.path.insert(1, repo_path) 
sys.path.insert(2, '/nfs/team205/lm25/thymus_projects/thymus_ageing_atlas/General_analysis/scripts')

# Define paths
plots_path = f'{repo_path}/plots/'
data_path = f'{repo_path}/data/'
```

```python
author_df = pd.read_excel(f'{data_path}/curated/authors.xlsx')
author_df["initials"] = author_df.apply(
    lambda r: (
        (r["First Name"][0] if pd.notna(r["First Name"]) and r["First Name"] != "" else "")
        + (r["M. Initial"][0] if pd.notna(r["M. Initial"]) and r["M. Initial"] != "" else "")
        + (r["Last Name"][0] if pd.notna(r["Last Name"]) and r["Last Name"] != "" else "")
    ),
    axis=1
)
author_df.loc[author_df['First Name'] == 'Lisa Marie', "initials"] = 'LMM'
```

```python
author_df
```

```python
# Explode contributions into rows
col = "Contributions (see https://groups.niso.org/higherlogic/ws/public/download/31067/CRediT_Taxonomy_Terms_and_Definitions_list)"
# split the contribution strings into lists, explode into rows, strip whitespace and convert empty strings to NaN
author_df[col] = author_df[col].fillna('').astype(str).str.split(',') 
author_df = author_df.explode(col)
author_df[col] = author_df[col].str.strip()
author_df.loc[author_df[col] == '', col] = np.nan

author_df
```

```python
author_df.groupby(col).agg({'initials': lambda x: ', '.join(x)}).reset_index().to_csv(f'{data_path}/curated/authors_contributions_exploded.csv', index = False)
```
